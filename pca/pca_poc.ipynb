This is a sample code to explore PCA using sklearn.

Exploratory Analysis of the iris dataset and sample pca implementation

```python
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from mpl_toolkits.mplot3d import Axes3D
from random import *
 
```

```python
iris = datasets.load_iris()

# columns are sepal length, width, petal length and width

X = iris.data[:,:2]
y = iris.target;
print (X.size,y.size);

x_min, x_max = X[:,0].min()-.5, X[:,0].max()+.5
y_min, y_max = X[:,1].min()-.5, X[:,1].max()+.5



plt.figure(2, figsize=(8, 6))
plt.clf()
```

```python
plt.scatter(X[:,0],X[:,1],c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())

plt.show()

```

```python
fig = plt.figure(1,figsize=(6,4))

ax = Axes3D(fig, elev=-150, azim=110)
X_reduced = PCA(n_components=3).fit_transform(iris.data)

ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y,
           cmap=plt.cm.Set1, edgecolor='k',s=40 )
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])
plt.show()
```

PCA on the iris dataset to convert it to 2D


```python
target_names = iris.target_names;

print (target_names)

pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)


plt.figure()

print(X_r.shape)
print(X_r.dtype)
    

colors=['navy', 'turquoise', 'darkorange']
lw=1

for color,i,target_name in zip(colors,[0,1,2],target_names):
    print(color,i,target_name)
    plt.scatter(X_r[y==i,0],X_r[y==i,1], color=color, alpha=.8, lw=lw,
                label=target_name) 
   
    
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')

plt.figure()
plt.show()
    
```
